```python
import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
```

#### 1. **Crear una función que devuelva los documentos *Tokenizados* (una lista de listas) y con los tokens (palabras) en minúsculas.**

* **input**: lista de documentos (lista de Strings).
* **output**: lista de listas, en la que cada lista contiene los tokens del documento.

```python
def tokenization(docs_list):
    tokens = []
    for doc in docs_list:
        lista_tokenizada = [word.lower() for text in doc for word in word_tokenize(text)]
        tokens.append(lista_tokenizada)
    return tokens

corpus = tokenization(corpus)
print(corpus)
```

#### 2. **Crear una función que elimine los tokens que sean signos de puntuación y *Stop-Words*.**

* **input**: lista de listas, en la que cada lista contiene los tokens del documento.
* **output**: lista de listas, en la que cada lista contiene los tokens del documento.

```python
# Crear un stopwords
"""a = set(stopwords.words("spanish"))"""
```

```python
# Ver la puntuación de un String
"""punctuation = set(string.punctuation)
print(punctuation)"""
```

```python
def remove_words(docs):
    stop_words = set(stopwords.words("spanish"))
    punctuation = set(string.punctuation)
    other_words = ["“", "”"]
    without_stop_words = []
    for doc in docs:
      without_stop_and_punct = [word for word in doc if word.lower() not in stop_words and word not in punctuation and word not in other_words]
      without_stop_words.append(without_stop_and_punct)

    return without_stop_words
corpus = remove_words(corpus)
print(corpus)
```

```python
def lematization(docs):
    #Forma 1 (NLTK)
    """pst = PorterStemmer()
    lem_result = []
    for doc in docs:
      lem = [pst.stem(token) for token in doc]
      lem_result.append(lem)
    return lem_result"""

    #Forma 2 (Spacy)
    lem_result = []

    for text in docs:
        doc_str = ' '.join(text)
        doc = nlp(doc_str)
        lemas = [token.lemma_ for token in doc]
        lem_result.append(lemas)

    return lem_result

corpus = lematization(corpus)
print(corpus)
print(len(corpus[0]))
```

#### 4. **Crear una función que elimine todos los tokens que no sean *Nombres* (NOUN, PROPN), *Verbos*, *Advervios* o *Adjetivos*.**

* **input**: lista de listas, en la que cada lista contiene los tokens del documento.
* **output**: lista de listas, en la que cada lista contiene los tokens del documento.

```python
# NKTL para filtrar
"""import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
nltk.download('averaged_perceptron_tagger')

# Oración de ejemplo
oracion = "The quick brown fox jumps over the lazy dog"

# Tokenizar la oración
palabras = word_tokenize(oracion)

# Etiquetado de habla
etiquetas_pos = pos_tag(palabras)
print(etiquetas_pos)
# Definir las etiquetas de partes del discurso que deseas mantener
etiquetas_permitidas = ['DT', 'JJ']

# Filtrar las palabras basadas en las etiquetas de partes del discurso permitidas
palabras_filtradas = [palabra for palabra, pos in etiquetas_pos if pos in etiquetas_permitidas]

# Imprimir los resultados
print("Oración original:", oracion)
print("Palabras filtradas:", " ".join(palabras_filtradas))"""
```

```python
def filter_words(docs):
    doc_filter = []

    for documento_tokens in docs:
        doc_str = ' '.join(documento_tokens)
        doc = nlp(doc_str)

        tok_filter = [token.lemma_ for token in doc if token.pos_ not in ['NOUN', 'PROPN', 'VERB', 'ADV', 'ADJ']]
        doc_filter.append(tok_filter)

    return doc_filter

corpus = filter_words(corpus)
print(corpus)
print(len(corpus[0]))
```

#### 5. **Función que dada una lista de documentos, devuelva los documentos normalizados. Este ejercicio ya esta hecho y simplemente tiene que funcionar llamando a las 4 funciones anteriores.**

* **input**: lista de documentos (lista de Strings).
* **output**: lista de listas, en la que cada lista contiene los tokens del documento normalizados.

```python
def normalization(docs_list):
    corpus = tokenization(docs_list)
    corpus = remove_words(corpus)
    corpus = lematization(corpus)
    corpus = filter_words(corpus)
    return corpus

corpus = normalization(corpus)
```

#Análisis de sentimiento

```python
import matplotlib.pyplot as plt
import networkx
from textblob import TextBlob
```

```python
# definimos una funcion para analizar los sentimientos en cada texto del tweet
def analize_sentiment(tweet):
    '''
    Función de utilidad para clasificar la polaridad de un tweet
     usando textblob.
    '''
    analysis = TextBlob(tweet)
    if analysis.sentiment.polarity > 0:
        return 1
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return -1
```

```python
# Creamos una columna con el resultado del análisis:
data['SA'] = np.array([ analize_sentiment(tweet) for tweet in data['text_cleaned'] ])

data.head()
```

```python
# Cree un df que contenga el valor de polaridad y el texto del tweet
sentiment_df = data[['text_cleaned', 'SA']]

sentiment_df.head(15)
```

# Conocimiento adquirido

```python
pos = len(sentiment_df[sentiment_df['SA'] > 0]. value_counts())
neu = len(sentiment_df[sentiment_df['SA'] == 0]. value_counts())
neg = len(sentiment_df[sentiment_df['SA'] < 0]. value_counts())
tot =  pos+neu+neg
#print (pos, neu, neg, pos+neu+neg)
```

```python
# Visualizamos los %:

print("Porcentaje de tweets positivos : {}%".format(round(pos/tot*100,2)))
print("Porcentaje de tweets neutrales : {}%".format(round(neu/tot*100,2)))
print("Porcentaje de tweets negativos : {}%".format(round(neg/tot*100,2)))
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

# Plot histogram of the polarity values
sentiment_df.hist(bins=[-1, -0.75, -0.5, -0.25, 0.25, 0.5, 0.75, 1],
             ax=ax,
             color="purple")

plt.title("Sentimentos de Tweets sobre Trump Banned en Twitter")
plt.show()
```
