```python
from minisom import MiniSom
import numpy as np
import matplotlib.pyplot as plt
```

```python
som = MiniSom(x = 2, y = 2, input_len = 2, sigma = 1.0, learning_rate = 0.5)
```

```python
weights = np.array([[[-0.2, 0.2, -1, 0.3],
                    [[0.2, -1.2, -1.2, 0.3], )
weights
```

```python
u_matrix = som.distance_map()
u_matrix
```

```python
# Plot the U-Matrix
plt.imshow(u_matrix, cmap='bone_r')
plt.colorbar(label='distance from neurons in the neighbourhood')
plt.title('U-Matrix')
plt.show()
```
